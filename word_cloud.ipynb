```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# texto com comentarios
comentarios = '''

Sou cliente a decadas do Bradesco. Me forcaram meu investimento para Agora.
minha gerente do Prime diz que assunto com Agora so atraves do chat.
no chat diz que para plataforma que devo fazer solicitacao e negam numero de protocolo.
Sinceramente. nao e verdade. Minha experiencia no bradesco sempre foi positiva, depois que

E viva santo Antônio, que ele interceda por todos nós, e nossas Famílias á Jesus AMÉM
Sou cliente a 18 anos e amo esse Banco .

Eu sou teu Bradesco pra mim e excelente qualidade



Aqui na minha cidade é um péssimo banco,pessimo atendimento os clientes ficam mais de 1 hr esperando atendimento

Deus abençoe Bradesco por ter dado oportunidade a um anão.obrigado.meu servo. sou Deus

Sou cliente a um tempinho e gosto  do que recebo

Maria Alice,ana clara Bruno e tassia que Deus abençoe a nossa noite e nossas vidas em nome de Jesus derramando muitas saúde a todos da nossa família.
'''
# lista de palavras indesejadas para a analise
iguinorar_texto = ['a', 'á', 'e', 'é', 'o', 'u', 'mas', 'um', 'uma', 'ele', 'ela', 'mais', 'eu', 'aqui', 'voce', 'do', 'da', 'de', 'por', 'sou', 'em']

#filtro das palavras legitimas para analise
tratar_comentario = [palavra for palavra in comentarios.split() if palavra.lower() not in iguinorar_texto]
comentarios = str(tratar_comentario)

# instanciando o objeto wordcloud
nuvem_palavras = WordCloud(
    width=1600,
    height=1800,
    background_color='black'
).generate(comentarios)

# gerando a figura
plt.figure(figsize=(26,18))
plt.imshow(nuvem_palavras, interpolation='bilinear')
plt.axis('off')
plt.show()
```
